# Analizando los datos (transformando de voz a texto)

```python
from deepgram import Deepgram
import json

# API key
DEEPGRAM_API_KEY = '917142653c3f9651777a464ddaf9b3cc86f4cd60'

# Replace with your file path and audio mimetype
PATH_TO_FILE = 'vozAIAEF.ogg'
MIMETYPE = 'audio/ogg'

def main():
    # Initializes the Deepgram SDK
    dg_client = Deepgram(DEEPGRAM_API_KEY)

    with open(PATH_TO_FILE, 'rb') as audio:
        source = {'buffer': audio, 'mimetype': MIMETYPE}
        options = { "punctuate": True, "model": "nova", "language": "es-419" }

        print('el archivo puede tardar algunos minutos en ser procesado')
        response = dg_client.transcription.sync_prerecorded(source, options)
        print(json.dumps(response, indent=4))

main()
```
